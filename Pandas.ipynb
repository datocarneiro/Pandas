# ***Métodos Uteis***


| Método               | Descrição                                                                                           |
|----------------------|-----------------------------------------------------------------------------------------------------|
| `read_csv()`         | Lê um arquivo CSV e retorna um DataFrame.                                                            |
| `to_csv()`           | Exporta um DataFrame para um arquivo CSV.                                                            |

| `describe()`         | Fornece estatísticas descritivas de colunas numéricas (média, desvio padrão, mínimo, etc.).           |
| `groupby()`          | Agrupa dados usando uma coluna ou índice como chave e aplica funções agregadas.                      |
| `merge()`            | Mescla dois DataFrames com base em uma ou mais colunas.                                              |
| `concat()`           | Concatena dois ou mais DataFrames ao longo de um eixo (linhas ou colunas).                           |
| `pivot_table()`      | Cria uma tabela dinâmica para sumarizar e agrupar dados.                                             |
| `drop()`             | Remove linhas ou colunas de um DataFrame.                                                            |
| `dropna()`           | Remove valores ausentes (NaN) de um DataFrame.                                                       |
| `fillna()`           | Preenche valores ausentes (NaN) com um valor específico.                                             |
| `apply()`            | Aplica uma função ao longo de um eixo (linhas ou colunas) do DataFrame.                              |
| `iloc[]`             | Seleciona dados com base em índices de posição.                                                      |
| `loc[]`              | Seleciona dados com base em rótulos de linha ou coluna.                                              |
| `sort_values()`      | Ordena os dados com base em valores de colunas específicas.                                          |
| `value_counts()`     | Retorna a contagem de valores únicos em uma coluna.                                                  |
| `isnull()`           | Retorna um DataFrame booleano indicando valores nulos.                                               |
| `duplicated()`       | Verifica se existem entradas duplicadas no DataFrame.                                                |
| `count()`            | Retorna o número de valores não nulos em cada coluna ou linha.                                       |
| `sum()`              | Retorna a soma dos valores de uma coluna ou DataFrame.                                               |
| `mean()`             | Calcula a média dos valores de uma coluna ou DataFrame.                                              |
| `min()`              | Retorna o valor mínimo de uma coluna ou DataFrame.                                                   |
| `max()`              | Retorna o valor máximo de uma coluna ou DataFrame.                                                   |
| `std()`              | Calcula o desvio padrão de uma coluna ou DataFrame.                                                  |
| `median()`           | Calcula a mediana dos valores de uma coluna ou DataFrame.                                            |
| `mode()`             | Retorna o(s) valor(es) mais frequente(s) de uma coluna.                                              |
| `cumsum()`           | Retorna a soma cumulativa dos valores ao longo de um eixo.                                           |
| `cumprod()`          | Retorna o produto cumulativo dos valores ao longo de um eixo.                                        |
| `nunique()`          | Retorna o número de valores únicos em uma coluna ou DataFrame.                                       |
| `idxmax()`           | Retorna o índice da primeira ocorrência do valor máximo.                                             |
| `idxmin()`           | Retorna o índice da primeira ocorrência do valor mínimo.                                             |
| `nlargest()`         | Retorna as n maiores entradas de uma coluna especificada.                                            |
| `nsmallest()`        | Retorna as n menores entradas de uma coluna especificada.                                            |
| `corr()`             | Calcula a correlação entre colunas numéricas.                                                        |
| `cov()`              | Calcula a covariância entre colunas numéricas.                                                       |
| `mad()`              | Retorna a média da diferença absoluta dos valores (média absoluta da dispersão).                     |
| `rank()`             | Atribui rankings aos valores dentro de uma coluna.                                                   |
| `sample()`           | Retorna uma amostra aleatória de linhas do DataFrame.                                                |
| `explode()`          | Transforma uma lista ou array de valores em várias linhas, uma por valor.                            |
| `between()`          | Retorna uma série booleana indicando se os valores estão entre dois valores especificados.           |
| `combine_first()`    | Mescla dois DataFrames, priorizando os valores não nulos do primeiro.                                |
| `melt()`             | Converte um DataFrame largo para o formato longo (ideal para visualização ou análise de dados).      |
| `memory_usage()`     | Retorna a quantidade de memória usada por cada coluna do DataFrame.                                  |
| `eval()`             | Avalia expressões no contexto de um DataFrame, acelerando operações com grandes datasets.            |
| `get_dummies()`      | Converte variáveis categóricas em variáveis dummy/indicadoras.                                       |
| `interpolate()`      | Preenche valores faltantes interpolando dados.                                                       |
| `merge_asof()`       | Mescla dois DataFrames com base na chave mais próxima (útil para séries temporais).                  |


### Criando um dataframe a partir de um dicionário

```python
# dataframe = pd.DataFrame()
venda = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }
vendas_df = pd.DataFrame(venda) 
```

### Visualização dos Dados
    - print
    - display

```python
display(vendas_df)
```

### Importando arquivos e bases de dados

```python
vendas_df = pd.read_excel("Vendas.xlsx") #para ler o arquivo usa read_excel ou a extensão do arquivo poderia ser csv por exemplo, e dentro dos parenteses o nome do arquivo ou o path dele 
display(vendas_df)
```

### Resumos de Visualização de Dados simples e úteis
    - head -  dentro do parenteses() coloca quantas linhas quer visualizar, nesse caso coloquei 10, nesse exemplo i´ra mostrar da linha 0 à 9
    - shape - retorna quantas linhas e colunas o arquivo tem
    - describe - retorna um resumo 

```python
display(vendas_df.head(10))
print(vendas_df.shape)
display(vendas_df.describe())
```

### Filtrando dados do dataframe, exemplo Pegar 1 coluna (e os pd.Series), uma coluna apenas podemos chamar de series, passando as colunas que queremos dentro dos [], cada coluna precisa ser aberto um colchetes[]

```python
produtos = vendas_df[['Produto', 'Numero de Ordem']]
display(produtos)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# pegar uma linha
display(vendas_df.loc[1]) #aqui i´ra retornar como textoi e não como tabela 
print('=*'*45)
display(vendas_df.loc[1:5]) #aqui ja retona como tabela

print('=*'*45)
# pegar linhas que correspondem a uma condição, nesse exemplo quiz que retornasse somente as vendas do produto camisetas 
vendas_de_camisetas = vendas_df.loc[vendas_df['Produto'] == 'Camiseta']
display(vendas_de_camisetas)

print('=*'*45)
# pegar linhas que correspondem a uma condição, nesse exemplo quiz que retornasse somente as vendas do produto camisetas, mas quero apenas algumas colunas em especifico, uso loc[linhas, colunas]
vendas_norteshopping_df = vendas_df.loc[vendas_df['Produto'] == 'Camiseta', ["Numero de origem", "Valor Unitário"]]
display(vendas_norteshopping_df)

print('=*'*45)
# pegar 1 valor específico, passando a linha e a coluna, nesse exmplo retornar linha 1 e o valor da coluna Produto
print(vendas_df.loc[1, 'Produto'])
```

### Adicionar 1 coluna (importANTE O ARQUIVO ORIGINAL SERÁ ALTERADO)

```python
# a partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)

# criar uma coluna com valor padrão
vendas_df.loc[:, "Imposto"] = 0  #loc[linha,coluna ]
display(vendas_df)
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
#display(vendas_dez_df)

vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

### Excluir linhas e colunas

```python
vendas_df = vendas_df.drop(0, axis=0) # axis=0 será o eixo para linha, axis=1 será o eixo para as colunas 
display(vendas_df)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# deletar linhas e colunas completamente vazias, onde não ha valor em nenhuma "celula"
#vendas_df = vendas_df.dropna(how='all', axis=1)

# deletar linhas que possuem pelo menos 1 valor vazio (se houiver apenas uma "celula" vazia irá excluir a linha inteira)
#vendas_df = vendas_df.dropna()

# preencher valores vazios
# preencher com a média da coluna
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean()) # mean = média,  aqui prencido os valores vazio com a media da coluna comissão 
display(vendas_df)

# preencher com o último valor
vendas_df = vendas_df.ffill()
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
# value counts = contar valores 
transacoes_loja = vendas_df['Produto'].value_counts()
display(transacoes_loja)
```

```python
# group by
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum() # aqui poderia ser media (mean) etc...
display(faturamento_produto)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro) aqui o cabeçalho deve ser identico

```python
gerentes_df = pd.read_excel("Gerentes.xlsx")
display(gerentes_df)

print('=*'*45)
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
